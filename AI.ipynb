```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Load dataset
_df = pd.read_csv("ai_workforce_displacement_global_2020_2026.csv")
df = _df.copy()
df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Improve display formatting
pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 120)

# Prepare dataset
df["year"] = df["year"].astype(int)
df["quarter"] = df["quarter"].astype(int)
df["quarter_year"] = df["year"].astype(str) + "Q" + df["quarter"].astype(str)
df["industry_sector"] = df["industry_sector"].astype("string")

display(df[[
    "country", "region", "income_group", "year", "quarter_label",
    "industry_sector", "sector_automation_risk_score", "ai_adoption_index",
    "pct_sector_workforce_displaced", "net_workforce_change_pct",
    "ai_tool_adoption_pct"
]].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")

numeric_cols = [
    "sector_automation_risk_score",
    "gdp_per_capita_usd",
    "ai_adoption_index",
    "pct_sector_workforce_displaced",
    "pct_sector_workforce_new_roles_created",
    "net_workforce_change_pct",
    "ai_skill_wage_premium_pct",
    "pct_workforce_female",
    "pct_displaced_roles_female",
    "reskilling_programs_count",
    "govt_ai_policy_score_1_to_10",
    "ai_tool_adoption_pct",
]

# Correlation heatmap
corr = df[numeric_cols].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix for AI and Workforce Metrics")
plt.tight_layout()
plt.show()

# Top 10 sectors by average displacement
sector_displacement = (
    df.groupby("industry_sector")["pct_sector_workforce_displaced"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
plt.figure(figsize=(10, 6))
sector_displacement.plot(kind="barh", color="#4C72B0")
plt.gca().invert_yaxis()
plt.xlabel("Average % Workforce Displaced")
plt.title("Top 10 Industry Sectors by Average Workforce Displacement")
plt.tight_layout()
plt.show()

# Trend of AI adoption and displacement by income group
trend_data = (
    df.groupby(["year", "income_group"])
    .agg(
        ai_adoption_index=("ai_adoption_index", "mean"),
        pct_displaced=("pct_sector_workforce_displaced", "mean"),
    )
    .reset_index()
)
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=trend_data,
    x="year",
    y="ai_adoption_index",
    hue="income_group",
    marker="o",
)
plt.title("Average AI Adoption Index by Income Group over Time")
plt.ylabel("AI Adoption Index")
plt.xlabel("Year")
plt.legend(title="Income Group", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=trend_data,
    x="year",
    y="pct_displaced",
    hue="income_group",
    marker="o",
)
plt.title("Average Workforce Displacement by Income Group over Time")
plt.ylabel("% Workforce Displaced")
plt.xlabel("Year")
plt.legend(title="Income Group", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# Automation risk vs displacement scatter
sample_df = df.sample(n=1500, random_state=42)
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=sample_df,
    x="sector_automation_risk_score",
    y="pct_sector_workforce_displaced",
    hue="income_group",
    size="ai_adoption_index",
    sizes=(20, 200),
    alpha=0.7,
)
plt.title("Automation Risk vs Workforce Displacement")
plt.xlabel("Sector Automation Risk Score")
plt.ylabel("% Workforce Displaced")
plt.legend(title="Income Group", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
